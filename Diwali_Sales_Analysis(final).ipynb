**Exploratory Data Analysis**

## Gender

```python
# plotting a bar chart for Gender and it's count

count = sns.countplot(x='Gender',data = df,hue='Gender',palette = "PiYG")
for bars in count.containers:
    count.bar_label(bars)
```

```python
df.groupby('Gender',as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
# plotting a bar chart for gender vs total amount

sales = df.groupby('Gender',as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
color = ("c","y")
sns.barplot(x='Gender',y = 'Amount',data = sales,hue ='Gender',palette=color)
plt.show()
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

## Age Group

```python
age_count = sns.countplot(x= "Age Group",data = df, hue = "Gender",palette = "rocket")

for bars in age_count.containers:
    age_count.bar_label(bars)
```

```python
df.groupby("Age Group", as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False)
```

```python
# Total Amount vs Age Group

sales_bar = df.groupby(["Age Group","Gender"], as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False)
sales_bars = sns.barplot(x="Age Group", y = "Amount", data = sales_bar,hue = "Gender")
```

## State

```python
df.groupby("State",as_index = False)["Orders"].sum().sort_values(by = "Orders", ascending = False)
```

```python
# total number of orders from top 10 states

order = df.groupby("State",as_index = False)["Orders"].sum().sort_values(by = "Orders", ascending = False).head(10)
plt.figure(figsize =(16,6))
sns.barplot(x = "State", y = "Orders", data = order,hue ="State",palette = "viridis")
plt.xticks(rotation=45)
plt.show()
```

```python
# total amount/sales from top 10 states

order = df.groupby("State",as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)
plt.figure(figsize =(16,6))
sns.barplot(x = "State", y = "Amount", data = order,hue ="State",palette = "flare")
plt.xticks(rotation=45)
plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

## Mritial status

```python
ax = sns.countplot(x="Marital_Status",data = df,hue ="Marital_Status", palette="YlGn")

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
ax = df.groupby(["Marital_Status","Gender"], as_index = False)["Amount"].sum().sort_values (by = "Amount", ascending = False)
sns.barplot(x="Marital_Status", y = "Amount", data = ax, hue = "Gender",palette="RdPu")
plt.show()
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

## Occupation

```python
plt.figure(figsize = (15,5))
ax = sns.countplot(x="Occupation",data = df,hue = "Occupation",palette = "muted")
plt.xticks(rotation = 45)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby("Occupation",as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
```

```python
ax = df.groupby("Occupation",as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
plt.figure(figsize=(16,5))
sns.barplot(x = "Occupation", y = "Amount",data = ax, hue ="Occupation",palette = "Set2")
plt.xticks(rotation = 45)
plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

```python
plt.figure(figsize = (16,5))
ax = sns.countplot(x = "Product_Category",data = df,hue = "Product_Category", palette = "Paired")
plt.xticks(rotation = 45)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = df.groupby("Product_Category",as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
plt.figure(figsize=(16,5))
sns.barplot(x = "Product_Category", y = "Amount",data = ax, hue ="Product_Category",palette = "plasma")
plt.xticks(rotation = 45)
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
